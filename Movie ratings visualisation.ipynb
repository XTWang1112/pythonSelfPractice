```python
import pandas as pd
import seaborn as sns
sns.set()
import matplotlib.pyplot as plt
```

```python
movies = pd.read_csv('movie.csv')
movies.head()
```

```python
movies.Genre.unique()
```

```python
movies.Studio.unique()
```

```python
genreFilter =['action', 'comedy', 'adventure', 'animation', 'drama']
studioFilter = ['Buena Vista Studios', 'Sony', 'Universal', 'WB', 'Paramount Pictures', 'Fox']

temp = movies[movies.Genre.isin(genreFilter)]
filteredMovie = temp[temp.Studio.isin(studioFilter)]

filteredMovie
```

```python
filteredMovie.Genre.unique()
```

```python
filteredMovie.Studio.unique()
```

```python
sns.set(style = 'darkgrid')

ax = sns.boxplot(data = filteredMovie, x = 'Genre', y = 'Gross % US', orient = 'v', color = 'lightgray', showfliers = False)
plt.setp(ax.artists, alpha = 0.5)

sns.stripplot(x='Genre', y='Gross % US', data=filteredMovie, jitter=True, size=6, linewidth=0, hue = 'Studio', alpha=0.7)

ax.axes.set_title('Domestic Gross % by Genre',fontsize=30)
ax.set_xlabel('Genre',fontsize=20)
ax.set_ylabel('Gross % US',fontsize=20)

ax.legend(bbox_to_anchor=(1.05, 1), loc=2)
plt.show()
```
